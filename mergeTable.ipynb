```python
df=pd.DataFrame()
```

```python
for i in range (1,20):
  response=requests.get("https://api.themoviedb.org/3/movie/top_rated?api_key=dd21ff3c575b7b9b794e0d810d1683e0&language=en-US&page={}".format(i))
  tempdf=pd.DataFrame(response.json()['results'])[['adult','title','id']]
  df=pd.concat([df,tempdf], ignore_index=True)
```

```python
df0 = df.head(50)
df0.head()
```
